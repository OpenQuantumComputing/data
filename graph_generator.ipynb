```python
# This notebook generates a dataset of connected random graphs

import networkx as nx
from graph_generators import *

K = [2, 4]
N = [5, 14]
I = range(5)
```

```python
# Generating Erdos-Renyi graphs
for k in K:
    for n in N:
        for  i in I:
            G = er(n, k, with_random_weights=True)
            #nx.write_gml(G, "sample_graphs/w_er_n{:d}_k{:d}_{:d}.gml".format(n, k, i))
nx.draw_networkx(G)
```

```python
# Generating Watts-Strogatz graphs
for k in K:
    for n in N:
        for  i in I:
            G = ws(n, k, with_random_weights=True)
            #nx.write_gml(G, "sample_graphs/w_ws_n{:d}_k{:d}_{:d}.gml".format(n, k, i))
nx.draw_networkx(G)
```

```python
# Generating Barabasi-Albert graphs
for k in K:
    for n in N:
        for  i in I:
            G = ba(n, k, with_random_weights=True)
            nx.write_gml(G, "sample_graphs/w_ba_n{:d}_k{:d}_{:d}.gml".format(n, k, i))
nx.draw_networkx(G)
```
